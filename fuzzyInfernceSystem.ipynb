```python
!pip install scikit-fuzzy
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
#input Variables
study_time = ctrl.Antecedent(np.arange(0, 21, 1), 'StudyTimeWeekly')  # 0 to 20 hours
absences = ctrl.Antecedent(np.arange(0, 31, 1), 'Absences')          # 0 to 30 days
parental_support = ctrl.Antecedent(np.arange(0, 5, 1), 'ParentalSupport')  # 0 to 4
```

```python
#output Variables
grade_class = ctrl.Consequent(np.arange(0, 5, 1), 'GradeClass')  # 0 to 4 (A to F)
```

```python
#membership functions for study Time Weekly -- using triangulat mf
study_time['Low'] = fuzz.trimf(study_time.universe, [0, 0, 6])
study_time['Moderate'] = fuzz.trimf(study_time.universe, [4, 10, 16])
study_time['High'] = fuzz.trimf(study_time.universe, [14, 20, 20])
```

```python
absences['Low'] = fuzz.trimf(absences.universe, [0, 0, 10])
absences['Moderate'] = fuzz.trimf(absences.universe, [8, 15, 22])
absences['High'] = fuzz.trimf(absences.universe, [20, 30, 30])
```

```python
#membership functions for parental_Support -- using triangular mf
parental_support['None'] = fuzz.trimf(parental_support.universe, [0, 0, 0.5])
parental_support['Low'] = fuzz.trimf(parental_support.universe, [0, 1, 2])
parental_support['Moderate'] = fuzz.trimf(parental_support.universe, [1, 2, 3])
parental_support['High'] = fuzz.trimf(parental_support.universe, [2, 3, 4])
parental_support['Very High'] = fuzz.trimf(parental_support.universe, [3, 4, 4])
```

```python
#membership functions for grade class -- using triangular mf --> output
grade_class['A'] = fuzz.trimf(grade_class.universe, [3.5, 4, 4])
grade_class['B'] = fuzz.trimf(grade_class.universe, [3, 3.25, 3.5])
grade_class['C'] = fuzz.trimf(grade_class.universe, [2.5, 2.75, 3])
grade_class['D'] = fuzz.trimf(grade_class.universe, [2, 2.25, 2.5])
grade_class['F'] = fuzz.trimf(grade_class.universe, [0, 1, 2])
```
